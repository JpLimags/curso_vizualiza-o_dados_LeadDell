```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10,20,25,5]
plt.bar(categorias,valores)
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Meu primeiro gr√°fico de barras')
plt.show()
```

```python
# Fazendo pequenas altera√ß~oes na cor e no tamanho da figura 

plt.figure(figsize=(9,3))
plt.bar(categorias, valores, color = 'darkred')
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Meu primeiro gr√°fico de barras')
plt.show()
```
