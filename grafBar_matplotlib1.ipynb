```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10,20,25,5]
plt.bar(categorias,valores)
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Meu primeiro gr√°fico de barras')
plt.show()
```

```python
# Fazendo pequenas altera√ß~oes na cor e no tamanho da figura 

plt.figure(figsize=(9,3))
plt.bar(categorias, valores, color = 'darkred')
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Meu primeiro gr√°fico de barras')
plt.show()
```

### Agora iremos analisar um conjunto de dados de filmes

```python
#importando o conjunto de dados

filmes = pd.read_json('https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json')
filmes.head()
```

```python
filmes.describe()
```

```python
#  Agrupando os dados buscando responder a seguinte pergunta, quantos filmes existem de cada g√™nero?

contagens = filmes.groupby(by = 'Genre').count()
contagens
```

```python
#Plotando um gr√°fico apartir dos dados apurados 
plt.figure(figsize = (8,5))
plt.bar(contagens.index, contagens['Worldwide_Gross_M'])
plt.show()
```

### Gerando gr√°fico de barras com o Seaborn

```python
sns.set(style ="whitegrid")
sns.countplot(x='Genre', data = filmes)
plt.show()
```

```python
# Agora iremos fazer a visualiza√ß√£o anterior corrigindo apenas a sobreposi√ß√£o do nome das barras

sns.set(style = "darkgrid")
plt.figure(figsize=(10,5))
plt.title('Quantidade de filmes por g√™neros')
sns.countplot(x = 'Genre', color='b', data=filmes)
plt.show()
```

```python
sns.set(style= 'darkgrid')
sns.countplot(y = 'Genre', color = 'b', data=filmes)
plt.show()
```

```python
#Apresentando a fun√ß√£o barplot

plt.figure(figsize=(10,5))
plt.title('Bilheteria m√©dia por g√™nero')
sns.barplot(x='Genre', y ='Worldwide_Gross_M', color = 'b', ci='sd', data  = filmes)
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.title("Bilherteria total por g√™nero")
sns.barplot(x='Genre', y = 'Worldwide_Gross_M', color='orange', estimator=sum, ci=None, data = filmes)
plt.show()
```
