```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8,5))
plt.title("Relação entre o valor da conta e gorjeta ")
sns.scatterplot(x="total_bill", y= "tip", data = tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Relação entre o valor da conta e valor da gorjeta")
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

#### Gráfico de dispersão para mais de duas variaveis

```python
plt.figure(figsize=(8,5))
plt.title("Relação entre valor da conta e alor da gorjeta pelo gênero")
sns.scatterplot(x="total_bill", y="tip", hue = 'sex', data=tips)
plt.show()
```

```python
#Aqui iremos demonstrar um gráfico de dispersão com muitas categorias 

plt.figure(figsize=(8,5))
plt.title('Relação entre o valor da conta e valor da gorjeta pelo dia da semana')
sns.scatterplot(x="total_bill", y = "tip", hue = "day", data = tips)
plt.show()
```

```python
#Utilizando a função replot

sns.relplot(x="total_bill", y="tip", col = "day", data = tips)
```

```python
#Utilizando a função regplot

sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```python
# Trabalhando com mais uma variável

plt.figure(figsize=(8,5))
plt.title("Relação entre conta e valor da conta, valor da gorjeta e tamanho da mesa")
sns.scatterplot(x='total_bill', y='tip', hue='size', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y= "tip", size= 'size', sizes=(20,200), data=tips)
plt.show()
```
