```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline
sns.set(style="darkgrid")
```

```python
hiv = pd.read_csv('https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/hiv_data.txt')
hiv
```

```python
plt.figure(figsize=(10,5))
plt.title('VariaÃ§Ã£o da taxa mÃ©dia de HIV global')
sns.lineplot(x='date', y='HIV Rate', data=hiv)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('VariaÃ§Ã£o da tacxa mÃ©ia de Hiv para os nÃ­veis de renda')
sns.lineplot(x='date', y='HIV Rate', hue='Income Level', ci = None,data = hiv)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("VariaÃ§Ã£o da taxa de HIV par cada paÃ­s")
sns.lineplot(x="date", y = "HIV Rate", hue= 'country', legend = False, data= hiv)
plt.show()
```
